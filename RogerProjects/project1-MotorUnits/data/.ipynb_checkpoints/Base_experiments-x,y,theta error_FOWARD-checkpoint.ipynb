```python
plt.figure(1, figsize=(16, 8))
# plt.plot(df1['time'], df1['y_error'],
#          linestyle=':', color='blue')
plt.scatter(df1['time'], df1['y_error'], s=1, label='x error (in the x^B direction of Roger)')
# plt.xlim([1.26,1.38])
# plt.ylim([-0.7,0.2])
plt.title('0.75 meters directly in front of Roger\n\
            Target = (0.010, -0.830, -1.559)', fontsize=25)
plt.xlabel('Time', fontsize=25)
plt.ylabel('Error', fontsize=25)
plt.axhline(y=0.0, color='y', linestyle=':')
plt.legend()
```
