```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('eye_K1_00000_B0_01789.txt', sep=',')
df2 = pd.read_csv('eye_K0_50000_B0_01265.txt', sep=',')
df3 = pd.read_csv('eye_K1_50000_B0_02191.txt', sep=',')
```

```python
plt.figure()
# plt.plot(df['time'], df[' theta__error'], df['time'], df[' theta_dot_error'])
plt.plot(df1['time'], df1[' theta__error'], 'r--',
         df2['time'], df2[' theta__error'], 'g--', 
        df3['time'], df3[' theta__error'], 'b--')
# plt.plot(df1['time'], df1[' theta_dot_error'], 'r--')
plt.title('K=1.0 B=0.0179')
plt.xlabel('time)')
plt.ylabel('theta_error')
plt.axis([1, 1.2, -0.2, 1.8])
# plt.figure(figsize=(1,1))
# plt.rcParams["figure.figsize"] = 30,20
# plt.rcParams.update({'font.size': 30})
plt.show()
```

```python
from mpl_toolkits.axes_grid1 import host_subplot
import mpl_toolkits.axisartist as AA
import matplotlib.pyplot as plt

host = host_subplot(111, axes_class=AA.Axes)
# plt.subplots_adjust(right=0.75)
plt.rcParams["figure.figsize"] = 15,8

par1 = host.twinx()
par2 = host.twinx()

# offset = 60
# new_fixed_axis = par2.get_grid_helper().new_fixed_axis
# par2.axis["right"] = new_fixed_axis(loc="right",
#                                     axes=par2,
#                                     offset=(offset, 0))

# par2.axis["right"].toggle(all=True)

# host.set_xlim(1, 1.2)
host.set_xlim(1, 1.2)
host.set_ylim(-0.2, 1.8)

host.set_xlabel("Time")
host.set_ylabel("Theta_Error")
# par1.set_ylabel("Temperature")
# par2.set_ylabel("Velocity")

p1, = host.plot(df1['time'], df1[' theta__error'], 'r--', label="K=1.00000 B=0.01789")
p2, = par1.plot(df2['time'], df2[' theta__error'], 'g--', label="K=0.50000 B=0.01265")
p3, = par2.plot(df3['time'], df3[' theta__error'], 'b--', label="K=1.50000 B=0.02191")

par1.set_ylim(-0.2, 1.8)
par2.set_ylim(-0.2, 1.8)

plt.title('Finding K and B')
host.legend()

# host.axis["left"].label.set_color(p1.get_color())
# par1.axis["right"].label.set_color(p2.get_color())
# par2.axis["right"].label.set_color(p3.get_color())

plt.draw()
plt.show()
```
