```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
params = {'legend.fontsize': 20,
          'legend.handlelength': 2}
plt.rcParams.update(params)
```

```python
# ARM
# K=30 B=6.5
df1 = pd.read_csv('CartesianError_FullyForwardStraight_K30_00000_B6_50000.txt', sep=',')
# df1.head(30)
```

```python
plt.figure(1, figsize=(16, 8))
plt.plot(df1['time'], df1['arm_x_y_error'],
         linestyle=':', label='Left Arm Error')
# plt.plot(df1['time'], df1['eye_theta_error_r'],
#          linestyle=':', label='Right Eye')
# plt.scatter(df1['time'], df1['eye_theta_error_l'], s=0.2, label='Left Eye')
# plt.scatter(df1['time'], df1['eye_theta_error_r'], s=0.2, label='Right Eye')
# plt.xlim([0.3,6.1])
# plt.ylim([-0.2,0.2])
plt.title('Cartesian Error for Arm K=30, B=6.5\n\
Stretching the Arm from original position to Fully Strectched Forward', fontsize=25)
plt.xlabel('Time (s)', fontsize=25)
plt.ylabel('Error (m)', fontsize=25)
plt.axhline(y=0.0, color='y', linestyle=':')
plt.legend()
plt.savefig('FromOriginToFullStretched.png')
```
