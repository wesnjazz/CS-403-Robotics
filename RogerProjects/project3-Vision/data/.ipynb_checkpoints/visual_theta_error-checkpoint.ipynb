```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
params = {'legend.fontsize': 20,
          'legend.handlelength': 2}
plt.rcParams.update(params)
```

# Three Displacements

```python
df1 = pd.read_csv('vis_three03K5.0B0.052.txt', sep=',')
df1.head(10)
```

```python
plt.figure(1, figsize=(16, 8))
# plt.plot(df1['time'], df1['eye_theta_error_l'],
#          linestyle=':', color='blue', label='Left Eye')
plt.scatter(df1['time'], df1['eye_theta_error_l'], s=5, label='Left Eye')
plt.scatter(df1['time'], df1['eye_theta_error_r'], s=5, label='Right Eye')
plt.xlim([0.11,0.42])
# plt.ylim([-0.7,1.0])
plt.title('*** Oculomotor error ***\n\
Target = three consecutive discrete placements\n\
in the order of 1.$-\pi/4$  2. Center 3. $\pi/4$ in base frame', fontsize=25)
plt.xlabel('Time', fontsize=25)
plt.ylabel('Error', fontsize=25)
plt.axhline(y=0.0, color='y', linestyle=':')
plt.legend()
plt.savefig('vis_three03_001.png')
```

```python
plt.figure(1, figsize=(16, 8))
# plt.plot(df1['time'], df1['eye_theta_error_l'],
#          linestyle=':', color='blue', label='Left Eye')
plt.scatter(df1['time'], df1['eye_theta_error_l'], s=20, label='Left Eye')
plt.scatter(df1['time'], df1['eye_theta_error_r'], s=20, label='Right Eye')
plt.xlim([0.11,0.18])
plt.ylim([-0.8,0.1])
plt.title('*** Oculomotor error ***\n\
Target = three consecutive discrete placements\n\
1.$-\pi/4$ in base frame ZOOMED IN', fontsize=25)
plt.xlabel('Time', fontsize=25)
plt.ylabel('Error', fontsize=25)
plt.axhline(y=0.0, color='y', linestyle=':')
plt.legend()
plt.savefig('vis_three03_002.png')
```
