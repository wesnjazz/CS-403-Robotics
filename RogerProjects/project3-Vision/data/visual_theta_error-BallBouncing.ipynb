```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
params = {'legend.fontsize': 20,
          'legend.handlelength': 2}
plt.rcParams.update(params)
```

```python
# EYE
# K=3.0 B=0.33
df1 = pd.read_csv('vis_bounce_3_0.33.txt', sep=',')
# df1.head(30)
```

```python
plt.figure(1, figsize=(16, 8))
plt.plot(df1['time'], df1['eye_theta_error_l'],
         linestyle=':', label='Left Eye')
plt.plot(df1['time'], df1['eye_theta_error_r'],
         linestyle=':', label='Right Eye')
# plt.scatter(df1['time'], df1['eye_theta_error_l'], s=0.2, label='Left Eye')
# plt.scatter(df1['time'], df1['eye_theta_error_r'], s=0.2, label='Right Eye')
plt.xlim([0.3,6.1])
plt.ylim([-0.2,0.2])
plt.title('*** Oculomotor error ***\n\
Bouncing Ball back and forth\n\
K=3, B=0.33', fontsize=25)
plt.xlabel('Time', fontsize=25)
plt.ylabel('Error', fontsize=25)
plt.axhline(y=0.0, color='y', linestyle=':')
plt.legend()
plt.savefig('vis_bounce_3_0.33.png')
```
