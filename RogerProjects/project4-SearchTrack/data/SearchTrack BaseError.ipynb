```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
params = {'legend.fontsize': 20,
          'legend.handlelength': 2}
plt.rcParams.update(params)
```

```python
df1 = pd.read_csv('P401.txt', sep=',')
```

```python
plt.figure(1, figsize=(16, 8))
plt.plot(df1['time'], df1['base_err'],
         linestyle=':', label='Base Error')
# plt.plot(df1['time'], df1['eye_theta_error_r'],
#          linestyle=':', label='Right Eye')
# plt.scatter(df1['time'], df1['base_err'], s=0.2, label='Base Error')
# plt.scatter(df1['time'], df1['eye_theta_error_r'], s=0.2, label='Right Eye')
# plt.xlim([0.3,6.1])
# plt.ylim([-0.2,0.2])
plt.title('*** Base Control Error ***', fontsize=25)
plt.xlabel('Time', fontsize=25)
plt.ylabel('Base Error', fontsize=25)
plt.axhline(y=0.0, color='y', linestyle=':')
plt.legend()
plt.savefig('P401.png')
```
