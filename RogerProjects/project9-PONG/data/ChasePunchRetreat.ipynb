```python
df1 = pd.read_csv('P9_chasepunchretreat.txt', sep=',')
```

```python
plt.figure(1, figsize=(16, 8))
# plt.plot(df1['time'], df1['base_heading'],
#          linestyle=':', label='Base heading')
plt.plot(df1['time'], df1['base_posX'],
         linestyle=':', label='Base position X', color='k')
# plt.plot(df1['time'], df1['base_posY'],
#          linestyle=':', label='Base position Y')
plt.plot(df1['time'], df1['ball_posX'],
         linestyle=':', label='Ball position X', color='red')
# plt.plot(df1['time'], df1['ball_posY'],
#          linestyle=':', label='Ball position Y')
plt.plot(df1['time'], df1['armLX'],
         linestyle=':', label='Left Arm position X', color='g')
# plt.plot(df1['time'], df1['armLY'],
#          linestyle=':', label='Left Arm position Y')
plt.plot(df1['time'], df1['armRX'],
         linestyle=':', label='Right Arm position X', color='b')
# plt.plot(df1['time'], df1['armRY'],
#          linestyle=':', label='Right Arm position Y')

# plt.scatter(df1['time'], df1['base_err'], s=0.2, label='Base Error')
# plt.scatter(df1['time'], df1['eye_theta_error_r'], s=0.2, label='Right Eye')
plt.xlim([-0.05,1.5])
plt.ylim([-3.0,1.5])
plt.title('*** ChasePunchRetreat() ***', fontsize=25)
plt.xlabel('Time', fontsize=25)
plt.ylabel('x/y/theta', fontsize=25)
plt.axhline(y=0.0, color='y', linestyle=':')
plt.legend(loc='best', handlelength=1, borderpad=0.0, labelspacing=0.5)
plt.savefig('P9_ChasePunchRetreat02.png')
```
