```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
params = {'legend.fontsize': 20,
          'legend.handlelength': 2}
plt.rcParams.update(params)
```

```python
df1 = pd.read_csv('P9_setpoint_on_X.txt', sep=',')
```

```python
df2 = pd.read_csv('P9_setpoint_off_X.txt', sep=',')
```

```python
plt.figure(1, figsize=(16, 8))
plt.plot(df1['time'], df1['base_heading'],
         linestyle=':', label='Base heading')
plt.plot(df1['time'], df1['base_posX'],
         linestyle=':', label='Base position X')
plt.plot(df1['time'], df1['base_posY'],
         linestyle=':', label='Base position Y')
plt.plot(df1['time'], df1['ball_posX'],
         linestyle=':', label='Ball position X')
plt.plot(df1['time'], df1['ball_posY'],
         linestyle=':', label='Ball position Y')

# plt.scatter(df1['time'], df1['base_err'], s=0.2, label='Base Error')
# plt.scatter(df1['time'], df1['eye_theta_error_r'], s=0.2, label='Right Eye')
# plt.xlim([0.3,6.1])
# plt.ylim([-1.0,2.0])
plt.title('*** WITH setpoint_filter() ***', fontsize=25)
plt.xlabel('Time', fontsize=25)
plt.ylabel('x/y/theta', fontsize=25)
plt.axhline(y=0.0, color='y', linestyle=':')
plt.legend()
plt.savefig('P9_setpoint_ON.png')
```

```python
plt.figure(1, figsize=(16, 8))
plt.plot(df2['time'], df2['base_heading'],
         linestyle=':', label='Base heading')
plt.plot(df2['time'], df2['base_posX'],
         linestyle=':', label='Base position X')
plt.plot(df2['time'], df2['base_posY'],
         linestyle=':', label='Base position Y')
plt.plot(df2['time'], df2['ball_posX'],
         linestyle=':', label='Ball position X')
plt.plot(df2['time'], df2['ball_posY'],
         linestyle=':', label='Ball position Y')

# plt.scatter(df1['time'], df1['base_err'], s=0.2, label='Base Error')
# plt.scatter(df1['time'], df1['eye_theta_error_r'], s=0.2, label='Right Eye')
# plt.xlim([0.3,6.1])
# plt.ylim([-1.0,2.0])
plt.title('*** WITHOUT setpoint_filter() ***', fontsize=25)
plt.xlabel('Time', fontsize=25)
plt.ylabel('x/y/theta', fontsize=25)
plt.axhline(y=0.0, color='y', linestyle=':')
plt.legend()
plt.savefig('P9_setpoint_OFF.png')
```
